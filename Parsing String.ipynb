```python
# @title
import sys
import io
import time
import numpy as np
```

```python
t = []
t1 = time.time()
def GetReverse(literal):
    if literal[0] == '~':
      return literal[1:]
    else:
        return '~' + literal

def is_satisfiable(n_vars, clauses):
    candidates = {frozenset()}

    for clause in clauses:
        temp = set()

        for s in candidates:
            for literal in clause:

                if GetReverse(literal) not in s:

                    temp.add(s | {literal})

        candidates = temp

        if len(candidates) == 0:
            return False

    return True


def load_case(f):
    n_vars, n_clauses = f.readline().split()
    clauses = [f.readline().strip().split(' v ') for _ in range(int(n_clauses))]
    return int(n_vars), clauses


t2 = time.time()
t.append(float(t2-t1))
def main(f=sys.stdin):
    num_cases = int(f.readline())
    n=0
    for i in range(num_cases):
        n_vars, clauses = load_case(f)
        print(clauses)
        n = n_vars
        result = is_satisfiable(n_vars, clauses)

        print(f"{'satisfiable' if result else 'unsatisfiable'}")
        print(f"Execution time: {sum(t)} seconds")
    return clauses, n
```

```python
data = """
1
3 5
X1 v X2 v X3
X1 v ~X2
X2 v ~X3
X3 v ~X1
~X1 v ~X2 v ~X3
""".strip()
```

```python
clauses,n  = main(io.StringIO(data))
```
